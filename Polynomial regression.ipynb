```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/Position_Salaries.csv")
x=data.iloc[:, 1:-1].values
y=data.iloc[:, -1].values
```

```python
print(x)
```

```python
print(y)
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x, y)
```

Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
x_poly=poly_reg.fit_transform(x)
reg2=LinearRegression()
reg2.fit(x_poly,y)
```

```python
plt.scatter(x,y, color='red')
plt.plot(x,reg.predict(x),color='blue')
plt.title('truth or false (linear regression)')
plt.x_label=("position level")
plt.y_label=("salary")
plt.show()
```

```python
plt.scatter(x,y, color='red')
plt.plot(x,reg2.predict(poly_reg.fit_transform(x)),color='blue')
plt.title('truth or false (polynomial regression)')
plt.x_label=("position level")
plt.y_label=("salary")
plt.show()
```

Linear Regression

```python
reg.predict([[6.5]])
```

Polynomial Regression

```python
reg2.predict(poly_reg.fit_transform([[6.5]]))
```
